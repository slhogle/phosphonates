# READ KEGG HIERARCHY

```r
path2ko <- readr::read_tsv('path2ko.tsv', col_names = c("PATH", "KO"))
pathdesc <- readr::read_tsv('path_desc.tsv', col_names = c("PATH", "DESC"))
```

# READ KO TERMS FOR SINGLE COMPLETE GENOME AND MAP TO KEGG PATHWAYS

```r
mygenome='SB-kegg-go.tsv'

GENOME_full <- readr::read_tsv(mygenome, col_names = c("ID", "DESC", "KO", "GO"))

GENOME_KO_list <- GENOME_full %>% 
    select(ID, KO) %>%
    separate_rows(KO) %>%
    mutate(dum="ko:") %>%
    unite(KO, dum, KO, sep='', remove=TRUE) %>% 
    mutate(KO=ifelse(KO=="ko:NA", NA, KO))

GENOME_KO_PATH <- left_join(GENOME_KO_list, path2ko) %>% 
    filter(str_detect(PATH, "map") | is.na(PATH)) %>% 
    left_join(., pathdesc)
```
