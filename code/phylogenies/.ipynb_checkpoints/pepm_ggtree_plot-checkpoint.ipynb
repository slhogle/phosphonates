```r
# load necessary libraries
library(tidyverse)
library(ggtree)
library(ape)
#library(phangorn)
```

```r
# load raxml file
pepm.ggtree <- read.raxml("RAxML_bipartitionsBranchLabels.pro_clust")
```

```r
pro <- fortify(pepm.ggtree) %>% 
    select(label) %>% 
    filter(str_detect(label, 'proch')) %>%
    mutate(group="prochlorococcus")
pel <- fortify(pepm.ggtree) %>% 
    select(label) %>% 
    filter(str_detect(label, 'pelagibac')) %>%
    mutate(group="pelagibacter")
euk <- fortify(pepm.ggtree) %>% 
    select(label) %>% 
    filter(str_detect(label, 'eukaryote')) %>%
    mutate(group="eukaryote")
```

```r
fortify(pepm.ggtree) %>% 
    select(label) %>% 
    filter(str_detect(label, ''))
```

```r
p <- ggtree(pepm.ggtree, layout="daylight") + 
    geom_nodepoint(aes(size=bootstrap)) +
    scale_size(range = c(0.1, 1), limits = c(50,100)) +
    geom_treescale() +
    theme(legend.position="right")
p %<+% bind_rows(pro, pel, euk) + geom_tippoint(aes(color=group))
ggsave("pepm_tree.png", width=7, height=7)
```

```r
# Clade Ia
Ia_node <- MRCA(sar11.ggtree, c("AG-349-L17", "HTCC9565"))
Ia.A_node <- MRCA(sar11.ggtree, c("HTCC1013", "HTCC9565")) ## cold water Ia clade
Ia.B_node <- MRCA(sar11.ggtree, c("HTCC7211", "AG-325-C10")) ## warm water Ia clade
Ia.C_node <- MRCA(sar11.ggtree, c("AG-390-D18", "HTCC9022")) ## HTCC8051, HTCC9022 clade
Ia.D_node <- MRCA(sar11.ggtree, c("AG-422-D23", "AG-349-L17")) ## HIMB083 clade
Ia.E_node <- MRCA(sar11.ggtree, c("AG-390-A02", "HIMB140")) ## HIMB140, HIMB122 clade
Ia.F_node <- MRCA(sar11.ggtree, c("AG-414-O11", "AG-430-E20")) ## Ia AG- clade I
Ia.G_node <- MRCA(sar11.ggtree, c("AG-325-J17", "AG-426-P20")) ## Ia AG- clade II
Ia.H_node <- MRCA(sar11.ggtree, c("HIMB1321", "HIMB5")) ## HIMB1321, HIMB5 clade
Ia.I_node <- MRCA(sar11.ggtree, c("BACL5_bin20", "AG-410-J13")) ## baltic clade

# Clade Ib.A
Ib.A_node <- MRCA(sar11.ggtree, c("AG-337-G04", "AG-426-M19"))
# Clade Ib.B
Ib.B_node <- MRCA(sar11.ggtree, c("AG-345-L07", "AG-365-D07"))
# Clade Ic
Ic_node <- MRCA(sar11.ggtree, c("AAA288-G21", "AG-464-B04"))
# Clade IIb
IIb_node <- MRCA(sar11.ggtree, c("ETNP-OMZ-SAG-A06", "AG-435-N18"))
# Clade IIa
IIa_node <- MRCA(sar11.ggtree, c("AG-325-E13", "AG-426-I15"))
IIa.A_node <- MRCA(sar11.ggtree, c("ETNP-OMZ-SAG-D09", "AG-426-I15"))
IIa.B_node <- MRCA(sar11.ggtree, c("AG-426-L23", "AG-345-F21"))
# Clade IIIa
IIIa_node <- MRCA(sar11.ggtree, c("AG-470-E16", "IMCC9063"))
# Clade IIIb
IIIb_node <- MRCA(sar11.ggtree, c("AAA027-J10", "AAA280-B11"))
# Clade IV
IV_node <- MRCA(sar11.ggtree, c("AG-333-C14", "AG-422-M16"))
IVa_node <- MRCA(sar11.ggtree, c("AG-333-C14", "AG-337-G06"))
IVb_node <- MRCA(sar11.ggtree, c("AG-422-B15", "AG-422-M16"))
# Clade V
V_node <- MRCA(sar11.ggtree, c("AG-349-C15", "B4_spades"))
Va_node <- MRCA(sar11.ggtree, c("AG-349-C15", "HIMB59"))
Vb_node <- MRCA(sar11.ggtree, c("AG-337-I02", "B4_spades"))
```

```r
p <- ggtree(sar11.ggtree, layout="circular") + 
  geom_nodepoint(aes(size=bootstrap)) +
  geom_treescale() +
  scale_size(range = c(0.1, 1), limits = c(50,100)) +
  geom_cladelabel(node=Ia_node, label="Ia", align=T, offset=.5) +
  geom_cladelabel(node=Ib.A_node, label="Ib.A", align=T, offset=.5) +
  geom_cladelabel(node=Ib.B_node, label="Ib.B", align=T, offset=.5) +
  geom_cladelabel(node=Ic_node, label="Ic", align=T, offset=.5) +
  geom_cladelabel(node=IIa_node, label="IIa", align=T, offset=.5) +
  geom_cladelabel(node=IIb_node, label="IIb", align=T, offset=.5) +
  geom_cladelabel(node=IIIa_node, label="IIIa", align=T, offset=.5) +
  geom_cladelabel(node=IIIb_node, label="IIIb", align=T, offset=.5) +
  geom_cladelabel(node=IV_node, label="IV", align=T, offset=.5) +
  geom_cladelabel(node=V_node, label="V", align=T, offset=.5) +
  geom_hilight(node=Ia_node, alpha=.4, fill="#F8766D") +
  geom_hilight(node=Ib.A_node, alpha=.4, fill="#D89000") +
  geom_hilight(node=Ib.B_node, alpha=.4, fill="#A3A500") +
  geom_hilight(node=Ic_node, alpha=.4, fill="#39B600") +
  geom_hilight(node=IIa_node, alpha=.4, fill="#00BFC4") +
  geom_hilight(node=IIb_node, alpha=.4, fill="#00BF7D") +
  geom_hilight(node=IIIa_node, alpha=.4, fill="#9590FF") +
  geom_hilight(node=IIIb_node, alpha=.4, fill="#00B0F6") +
  geom_hilight(node=IV_node, alpha=.4, fill="#E76BF3") +
  geom_hilight(node=V_node, alpha=.4, fill="#FF62BC") 
  #geom_tiplab(size=1, align=T) +
  #theme_tree2(legend.position='right') 

p.a <- flip(p, 231, 237)
p.b <- flip(p.a, 179, 188)
p.c <- flip(p.b, 154, 157)
```
